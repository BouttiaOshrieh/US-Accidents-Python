#### 2.Data cleaning

```python
df_primary.head(10)
```

```python
print(df_primary.isnull().sum())
```

```python
#creating a subset of the data to find common missing values between columns and remove them
df_clean = df_primary.dropna(subset = ['Wind_Chill(F)'])
```

```python
df_clean.isnull().sum()
```

```python
#creating new columns to extract date and year
df_clean['Date'] = df_clean['Weather_Timestamp'].str.split(' ').str[0]
```

```python
df_clean['Date'] = pd.to_datetime(df_clean['Date'])
```

```python
df_clean['Year'] = df_clean['Date'].dt.year
```

```python
df_clean['Year'] = df_clean['Year'].astype('int')
```

```python
df_clean
```

```python
#checking for duplicates
duplicates = df_clean[df_clean.duplicated()]
```

```python
duplicates
```

```python
#checking for mixed type columns
for col in df_clean.columns.tolist():
  weird = (df_clean[[col]].applymap(type) != df_clean[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_clean[weird]) > 0:
    print (col)
```

```python
#fixing mixed type columns
df_clean['City'] = df_clean['City'].astype('str')
```

```python
df_clean['Street'] = df_clean['Street'].astype('str')
```

```python
df_clean['Zipcode'] = df_clean['Zipcode'].astype('str')
```

```python
df_clean['Airport_Code'] = df_clean['Airport_Code'].astype('str')
```

```python
df_clean['Timezone'] = df_clean['Timezone'].astype('str')
```

```python
df_clean['Weather_Condition'] = df_clean['Weather_Condition'].astype('str')
```

```python
df_clean['Wind_Direction'] = df_clean['Wind_Direction'].astype('str')
```

```python
df_clean['Sunrise_Sunset'] = df_clean['Sunrise_Sunset'].astype('str')
```

```python
df_clean['Civil_Twilight'] = df_clean['Civil_Twilight'].astype('str')
```

```python
df_clean['Nautical_Twilight'] = df_clean['Nautical_Twilight'].astype('str')
```

```python
df_clean['Astronomical_Twilight'] = df_clean['Astronomical_Twilight'].astype('str')
```

```python
#checking for mixed type columns
for col in df_clean.columns.tolist():
  weird = (df_clean[[col]].applymap(type) != df_clean[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_clean[weird]) > 0:
    print (col)
```

```python
#changing Severtity and Number data types to string since those are not numerical data
df_clean['Severity'] = df_clean['Severity'].astype('str')
```

```python
df_clean['Number'] = df_clean['Number'].astype('str')
```
